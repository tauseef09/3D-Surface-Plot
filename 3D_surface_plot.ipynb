```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from matplotlib import cm
```

```python
df = pd.read_csv('Data-in-xyz-format-converted.csv')
df.head()
```

```python
# Obtaining the x,y,z values
x = df['Row']
y = df['Column']
z = df['value']
```

```python
# 3D surface plot in viridis colormap
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(x, y, z, cmap='viridis', edgecolor='none')
plt.show()
```

```python
# 3D surface plot in coolwarm colormap
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(x, y, z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
plt.show()
```
